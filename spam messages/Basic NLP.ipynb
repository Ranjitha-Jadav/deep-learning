# Pre-processing Text Data

Cleaning up the text data is necessary to highlight attributes that you're going to want your machine learning system to pick up on. We will explore three pre-processing steps in this lesson:
1.Remove punctuation
2.Tokenization
3.Remove stopwords


```python
msg.head()
```

# Remove Punctuation

```python
import string
string.punctuation
```

```python
def remove_punct(text):
    text="".join([char for char in text if char not in string.punctuation])
    return text
msg['text_clean']=msg['Text'].apply(lambda x:remove_punct(x))
msg.head()
```

# Tokenize

```python
# Define a function to split our sentences into a list of words
import re

def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens

msg['text_tokenized'] = msg['text_clean'].apply(lambda x: tokenize(x.lower()))

msg.head()
```

# Remove stopwords

```python
# Load the list of stopwords built into nltk
import nltk

stopwords = nltk.corpus.stopwords.words('english')
```

```python
# Define a function to remove all stopwords
def remove_stopwords(tokenized_text):    
    text = [word for word in tokenized_text if word not in stopwords]
    return text

msg['text_nostop'] = msg['text_tokenized'].apply(lambda x: remove_stopwords(x))

msg.head()
```
